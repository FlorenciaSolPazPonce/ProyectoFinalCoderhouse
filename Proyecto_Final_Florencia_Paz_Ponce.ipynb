```python
X = diab_procesado.drop(columnas_a_eliminar, axis=1)
y = diab_procesado['DiabetesColumna']
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

modelo = SVC()


num_folds = 5

metricas = ['accuracy', 'precision_macro', 'recall_macro', 'f1_macro']

for metrica in metricas:
    resultados = cross_val_score(modelo, X_test, y_pred, cv=num_folds, scoring=metrica)


    for i in range(num_folds):
        print(f"{metrica} - Fold {i+1}: {resultados[i]}")

    print(f"{metrica} - Resultado promedio: {np.mean(resultados)}")
    print()
```

***InterpretaciÃ³n de los resultados para cada mÃ©trica:***

**Accuracy:** Es la mÃ©trica mÃ¡s comÃºn para problemas de clasificaciÃ³n. Mide la proporciÃ³n de predicciones correctas sobre el total de predicciones. En este caso, el resultado promedio de accuracy es aproximadamente 0.927, lo que significa que el modelo estÃ¡ acertando alrededor del 92.7% de las veces en las predicciones.

**Precision_macro:** Esta mÃ©trica mide el promedio de precisiÃ³n para todas las clases en el problema de clasificaciÃ³n. La precisiÃ³n se refiere a la proporciÃ³n de predicciones positivas que son realmente correctas. El resultado promedio de precision_macro es aproximadamente 0.928, lo que indica que el modelo tiene una buena precisiÃ³n en las predicciones.

**Recall_macro:** Mide el promedio de recall para todas las clases en el problema de clasificaciÃ³n. Recall se refiere a la proporciÃ³n de ejemplos positivos que fueron correctamente identificados por el modelo. El resultado promedio de recall_macro es aproximadamente 0.925, lo que significa que el modelo tiene una buena capacidad para encontrar ejemplos positivos.

**F1_macro:** Esta mÃ©trica es el promedio armÃ³nico de precisiÃ³n y recall (ponderando igualmente ambas mÃ©tricas). El resultado promedio de f1_macro es aproximadamente 0.926, lo que indica que el modelo tiene un buen equilibrio entre precisiÃ³n y recall.

**En general, los resultados obtenidos indican que el modelo tiene un rendimiento sÃ³lido en tÃ©rminos de precisiÃ³n, recall, F1-score y accuracy.**
